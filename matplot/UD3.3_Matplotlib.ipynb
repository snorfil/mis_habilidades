### Visualización de la información a través de MatplotLib

Para trabajar con Matplot lo primero que haremos es importarla de matplotlib.pyplot

```python
import matplotlib.pyplot as plt

# importaremos la librería de panda
import pandas as pd

# habrá ejemplos de uso de la librería numPy
import numpy as np
```

```python
ts = pd.Series(np.random.randn(5))
print(ts)
```

## Diagrama de dispersión o puntos

```python
# se puede visualizar una Serie de forma simple, uniendo los puntos con una línea

plt.plot(ts)
plt.show()     #mostrar el grafico con la configuración anterior
```

```python
#pero también podemos visualizarlo como un diagrama de puntos
plt.plot(ts, 'o')
plt.show()
```

```python
#se puede realizar las dos cosas, crear un diagrama de puntos unidos con lineas
plt.plot(ts, marker='o')
plt.show()
```

```python
#puedes jugar con el formato de los puntos y de la línea siguiendo la documentación del siguiente enlace:
# https://www.w3schools.com/python/matplotlib_markers.asp
```

En un mismo diagrama se pueden mostrar varias gráficas. Por ejemplo:

```python
ts2 = pd.Series(np.random.randn(5))
plt.plot(ts, marker='*')
plt.plot(ts2, marker='o')
plt.show()
```

También se pueden añadir etiquetas a los ejes y un título al gráfico

```python
ts2 = pd.Series(np.random.randn(5))

plt.plot(ts, marker='*')
plt.plot(ts2, marker='o')

plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.title("Titulo gráfico")


plt.show()
```

Otra característica importante es añadir leyenda, para eso debemos incluir el atributo label en la función de plot. Posteriormente indicaremos su localización llamando a legend. 

```python
ts2 = pd.Series(np.random.randn(5))

plt.plot(ts, marker='*', label='ts')
plt.plot(ts2, marker='o', label='ts2')

plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.title("Titulo gráfico")

plt.legend(loc='best')


plt.show()
```

A veces, es recomendable para ver mejor los valores de las distribuciones añadir regilla. 

```python

plt.plot(ts, marker='*', label='ts')
plt.plot(ts2, marker='o', label='ts2')

plt.xlabel('Eje x')
plt.ylabel('Eje y')

plt.title("Titulo gráfico")

plt.legend(loc='best')

plt.grid() #se puede añadir atributo axis = 'x' para solo el eje X


plt.show()
```

# Diagramas de barras

```python
# también se puede visualizar un Dataframe de una forma muy simple
df1 = pd.DataFrame({ 
    'Name': ['John', 'Sammy', 'Joe'], 
    'Age': [45, 38, 90], 
    'Height(in cm)': [150, 180, 160] 
}) 
print(df1)
```

```python
df1.plot.bar()  #diagrama de barras vertical
plt.show()
```

Al llamar a plot podemos indicar las etiquetas que tendrá el eje X y el eje Y

```python
df1.plot.bar(x="Name", y=["Age", "Height(in cm)"]) 
plt.show()
```

Este mismo diagrama lo podemos poner en horizontal

```python
df1.plot.barh(x="Name", y=["Age", "Height(in cm)"]) 
plt.show()
```

```python
# puedes probar diferentes estilos en los diagramas de barras siguiendo el tutorial del enlace:
# https://www.w3schools.com/python/matplotlib_bars.asp
```

## Diagrama de áreas

```python
# existen datos que se ven mejor en forma de área
df1.plot.area()
plt.grid(True)
plt.show()
```

#### Mapas de calor

Para generar un mapa de calor utilizaremos la libreria seaborn

```python
# muchas veces es muy util un heatmap
import seaborn as sns

ventas = pd.DataFrame(
    np.random.randint(1, 100, 120).reshape(12, 10),
    columns = np.arange(2009, 2019),
    index = ["ene", "feb", "mar", "abr", "may", "jun", "jul", "ago", "sep", "oct", "nov", "dic"]
)
ventas.describe()
```

```python
sns.heatmap(ventas, annot=True) 
```

Podemos especificar un mapa de color con el parámetro cmap, al que podemos asignar el nombre de una paleta de colores:

```python
sns.heatmap(ventas, annot=True, cmap="PiYG")
```

```python
sns.heatmap(ventas, annot=True, cmap="Blues")
```

## Diagrama de quesos

```python
# si deseamos realizar un diagrama de quesos

ventas.loc['ene'].plot.pie(subplots=True)
plt.show()
```

# Creando histogramas

Los histogramas nos sirven para mostrar la frecuencia de las distribuciones. Nos mostrará el número de observaciones por cada intervalo. 

```python
x = np.random.normal(170, 10, 250) #crearemos una distribución normal.
print(x)
```

```python
plt.hist(x)    
plt.show()
```
